# üìä **Complete EDA Notebook (Maxed-Out Version)**
This notebook performs an extensive Exploratory Data Analysis (EDA) with deep insights, interpretation, and visualizations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (12,6)

df = pd.read_csv('train_dataset.csv')
df.head()
```

## üîç **Dataset Overview**

```python
df.info()
```

### üìù **Insights**
- No missing values means the dataset is clean.
- All columns are numeric.

## üß™ **Missing Values Check**

```python
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

### üìù **Insights**
- No missing values detected.
- No imputation strategies needed.

## üìà **Statistical Summary**

```python
df.describe().T
```

### üìù **Insights**
- Very wide range for liver enzymes (AST, ALT, GTP) indicates heavy right-skew.
- Waist, weight, triglycerides show strong variation ‚Üí important for metabolic health.
- HDL and LDL levels suggest possible dyslipidemia trends.

## üî¢ **Unique Values Per Column**

```python
df.nunique()
```

### üìù **Insights**
- Indicates which columns are continuous vs categorical.
- `hearing`, `dental caries`, `Urine protein` behave like categorical variables.

## üö¨ **Target Variable Distribution (Smoking)**

```python
plt.figure(figsize=(6,4))
sns.countplot(x=df['smoking'])
plt.title('Smoking Distribution')
plt.show()

# Statistics table
stats = pd.DataFrame({
    "Count": df['smoking'].value_counts(),
    "Percentage (%)": df['smoking'].value_counts(normalize=True) * 100
})

print("Smoking Statistics:\n")
print(stats)
```

### üìù **Insights**
- Slight class imbalance: more non-smokers.
- Consider stratified splits.
- Normal imbalance for medical datasets.

## üìä **Feature Distributions (Histograms)**

```python
df.hist(bins=40, figsize=(18,15))
plt.tight_layout()
plt.show()
```

### üìù **Insights**
- Most physiological and biochemical metrics show right-skew.
- Outliers visible in liver markers ‚Üí common in heavy drinking/smoking populations.
- Height, age show normal-ish distribution.

## üî• **Smoker vs Non-Smoker Distribution Comparison**

```python

cols = ['waist(cm)','triglyceride','HDL','LDL','AST','ALT','Gtp']

for c in cols:
    plt.figure(figsize=(10,5))
    sns.kdeplot(df[df.smoking==0][c], shade=True, label='Non-Smoker')
    sns.kdeplot(df[df.smoking==1][c], shade=True, label='Smoker')
    plt.title(f'KDE of {c} Distribution by Smoking Status')
    plt.legend()
    plt.show()
```

### üìù **Insights**
- Smokers show higher waist, ALT, AST, GTP ‚Üí strong liver stress pattern.
- HDL lower in smokers (expected ‚Äî smoking reduces good cholesterol).
- Triglycerides significantly higher in smokers.

## üîó **Correlation Heatmap**

```python

plt.figure(figsize=(16,10))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

### üìù **Insights**
- Strong collinearity in liver enzymes (AST/ALT/GTP cluster).
- Weight ‚Üî waist ‚Üî triglycerides cluster.
- HDL is inversely correlated with triglycerides.
- Smoking correlates with GTP, triglycerides, ALT, AST, waist.

## üëÄ Pairplot of Important Features

```python

sns.pairplot(df[['waist(cm)','triglyceride','HDL','ALT','smoking']], hue='smoking')
plt.show()
```

### üìù **Insights**
- Clustering clearly separates smokers and non-smokers.
- Waist + triglyceride + ALT create a strong smoking signature.

## ‚ö†Ô∏è **Outlier Detection (Boxplots)**

```python

num_cols = df.select_dtypes(include=np.number).columns

for c in ['AST','ALT','Gtp','triglyceride','waist(cm)']:
    sns.boxplot(x=df[c])
    plt.title(f'Boxplot: {c}')
    plt.show()
```

### üìù Insights
- Heavy right-tail outliers confirm typical clinical distribution.
- High GTP outliers often correlate with smoking or alcohol consumption.
- Outliers not to be removed ‚Äî medically meaningful.
